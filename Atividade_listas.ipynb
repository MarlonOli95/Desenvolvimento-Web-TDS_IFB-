```python
#Faça um programa que leia n notas, mostre as notas e a média.
soma = 0
notas = []
i = 1
n = int(input("Quantas notas seriam calculadas?: "))
for i in range (n):
 dado = float(input("Qual a "+str(i + 1)+"° nota?: "))
 notas.append(dado)
 print(notas)

for i in range (n):
  soma = soma + notas[i]

media = soma/n
print(media)
```
