```python
'''
TAREFA 5
Faça um programa que solicita uma
quantidade indefinida de números positivos
(deve parar quando um número negativo for
digitado).
Armazene-os em duas listas:
uma para os números pares;
outra para os números ímpares.
A seguir, mostre a porcentagem de pares e
ímpares digitados.
'''
NumeroPar = []
NumeroImpar = []
#O primeiro numero não será retornado.
numero = int(input("Digite um numero: "))

while numero > 0:
  numero = int(input("Digite um numero: "))
  if numero%2 != 0:
    NumeroImpar.append(numero)
  if numero%2 == 0:
    NumeroPar.append(numero)
  if numero<0:
    if numero%2 == 0:
     NumeroPar.remove(numero)
    if numero%2 != 0:
     NumeroImpar.remove(numero)
     break

print("Lista dos números pares: ",NumeroPar)
print("Lista dos números impares: ",NumeroImpar)
```
